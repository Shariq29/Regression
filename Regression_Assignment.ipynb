## Q NO 1: What is Simple Linear Regression ?

## ANS:

Simple Linear Regression is a statistical method used to model the relationship between two variables: one independent variable (predictor) and one dependent variable (response). It assumes a linear relationship, meaning the change in the dependent variable is proportional to the change in the independent variable. The goal is to find the best-fitting straight line that predicts the dependent variable based on the independent variable.

The equation of the line is:

\[ y = \beta_0 + \beta_1x + \epsilon \]

- \( y \): Dependent variable (what you’re predicting)
- \( x \): Independent variable (the predictor)
- \( \beta_0 \): Intercept (the value of \( y \) when \( x = 0 \))
- \( \beta_1 \): Slope (the change in \( y \) for a unit change in \( x \))
- \( \epsilon \): Error term (captures random variation not explained by the model)

### Key Points:
1. **Purpose**: Predict \( y \) given \( x \), or understand the relationship between them.
2. **Assumptions**:
   - The relationship between \( x \) and \( y \) is linear.
   - Errors (\( \epsilon \)) are normally distributed with a mean of 0.
   - Variance of errors is constant (homoscedasticity).
   - Observations are independent.
3. **Fitting the Line**: The best-fitting line minimizes the sum of squared differences between the actual and predicted \( y \)-values (using a method like Ordinary Least Squares).
4. **Output**:
   - Slope (\( \beta_1 \)): Indicates the strength and direction of the relationship.
   - Intercept (\( \beta_0 \)): The starting point of the line.
   - \( R^2 \): A measure of how much of the variability in \( y \) is explained by \( x \) (ranges from 0 to 1).

### Example:
Suppose you want to predict house prices (\( y \)) based on their size in square feet (\( x \)). Simple Linear Regression would find a line like:

\[ \text{Price} = 50,000 + 200 \cdot \text{Size} \]

Here, \( \beta_0 = 50,000 \) (base price), \( \beta_1 = 200 \) (price increase per square foot), early retirement and the model predicts prices based on size.

This method is widely used in fields like economics, science, and machine learning for its simplicity and interpretability.











## Q NO 2: What are the key assumptions of Simple Linear Regression ?

## ANS:

The key assumptions of Simple Linear Regression are:

1. **Linearity**: The relationship between the independent variable (\( x \)) and the dependent variable (\( y \)) is linear, meaning \( y \) changes proportionally with \( x \).

2. **Independence**: Observations are independent of each other. The value of \( y \) for one observation does not influence or depend on the value of \( y \) for another.

3. **Homoscedasticity**: The variance of the error terms (\( \epsilon \)) is constant across all values of \( x \). This means the spread of residuals (differences between actual and predicted \( y \)) is consistent.

4. **Normality of Errors**: The error terms (\( \epsilon \)) are normally distributed with a mean of 0. This assumption is particularly important for hypothesis testing and confidence intervals.

5. **No or Minimal Multicollinearity**: In simple linear regression, since there’s only one independent variable, this assumption is not applicable, but it becomes relevant in multiple linear regression.

6. **No Significant Outliers**: The data should not contain extreme outliers that could disproportionately influence the regression line.

These assumptions ensure the model is valid and its predictions are reliable. Violations (e.g., non-linearity or heteroscedasticity) may require data transformation or a different modeling approach.



## Q NO 3: What is heteroscedasticity, and why is it important to address in regression models ?

## ANS:

**Heteroscedasticity** refers to the situation in regression analysis where the variance of the error terms (\( \epsilon \)) is not constant across all values of the independent variable (\( x \)). In other words, the spread or variability of the residuals (differences between actual and predicted values) changes as \( x \) changes. This violates one of the key assumptions of Simple Linear Regression (and other regression models), which assumes **homoscedasticity** (constant variance of errors).

### Characteristics of Heteroscedasticity
- **Visual Representation**: In a residual plot (residuals vs. predicted values or \( x \)), heteroscedasticity often appears as a funnel shape, where residuals spread out more at certain levels of \( x \) or predicted \( y \).
- **Example**: In predicting house prices based on size, heteroscedasticity might occur if the variability in prices is much larger for larger houses than for smaller ones.

### Why Heteroscedasticity is Important to Address
1. **Biased Standard Errors**:
   - Heteroscedasticity leads to unreliable standard errors for the regression coefficients (\( \beta_0 \), \( \beta_1 \)). This affects the calculation of t-statistics and p-values, potentially leading to incorrect conclusions about the significance of predictors.
   
2. **Inefficient Estimates**:
   - While the regression coefficients (\( \beta_0 \), \( \beta_1 \)) remain unbiased in the presence of heteroscedasticity, they are no longer efficient (i.e., they have higher variance than necessary). This reduces the precision of the model.

3. **Invalid Hypothesis Tests and Confidence Intervals**:
   - Since standard errors are incorrect, hypothesis tests (e.g., testing if \( \beta_1 = 0 \)) and confidence intervals for the coefficients become unreliable, potentially leading to misleading inferences.

4. **Poor Model Fit and Predictions**:
   - Heteroscedasticity indicates that the model may not adequately capture the underlying data structure, leading to suboptimal predictions, especially in regions where variance is high.

### Addressing Heteroscedasticity
To mitigate heteroscedasticity, you can:
1. **Transform the Data**:
   - Apply transformations to the dependent variable (\( y \)), such as logarithmic, square root, or reciprocal transformations, to stabilize variance.
   - Example: If predicting house prices, use \( \log(\text{Price}) \) instead of raw Price.

2. **Use Weighted Least Squares (WLS)**:
   - Assign weights to observations based on the inverse of their variance to give less influence to observations with higher variability.

3. **Robust Standard Errors**:
   - Use heteroscedasticity-robust standard errors (e.g., White’s standard errors) to correct standard errors without altering the model. This ensures valid hypothesis tests and confidence intervals.

4. **Model Specification**:
   - Check if the model is misspecified (e.g., missing important predictors or non-linear terms). Adding relevant variables or using a non-linear model may resolve heteroscedasticity.

5. **Use a Different Model**:
   - If heteroscedasticity persists, consider alternative models like generalized linear models or non-parametric regression that don’t assume constant variance.

### Detection
- **Residual Plots**: Plot residuals against predicted values or \( x \). A non-random pattern (e.g., funnel shape) indicates heteroscedasticity.
- **Statistical Tests**: Use tests like the Breusch-Pagan test or White’s test to formally detect heteroscedasticity.

### Why It Matters
Addressing heteroscedasticity ensures the reliability of statistical inferences and improves the model’s predictive accuracy. Ignoring it can lead to misleading conclusions, such as overestimating or underestimating the significance of predictors, which is critical in fields like economics, finance, or social sciences where regression models guide decision-making.


















## Q NO 4: What is Multiple Linear Regression ?

## ANS:

**Multiple Linear Regression** is an extension of Simple Linear Regression that models the relationship between one dependent variable (response) and two or more independent variables (predictors). It is used to predict or understand how multiple factors influence a single outcome, assuming a linear relationship between the predictors and the response.

### Model Equation
The equation for Multiple Linear Regression is:

\[ y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_nx_n + \epsilon \]

- \( y \): Dependent variable (what you’re predicting)
- \( x_1, x_2, \dots, x_n \): Independent variables (predictors)
- \( \beta_0 \): Intercept (value of \( y \) when all \( x_i = 0 \))
- \( \beta_1, \beta_2, \dots, \beta_n \): Coefficients representing the change in \( y \) for a one-unit change in each \( x_i \), holding other predictors constant
- \( \epsilon \): Error term (captures unexplained variation)

### Key Features
1. **Purpose**:
   - Predict \( y \) based on multiple predictors.
   - Quantify the impact of each predictor while controlling for others.
   - Example: Predicting house prices (\( y \)) using size (\( x_1 \)), number of bedrooms (\( x_2 \)), and location score (\( x_3 \)).

2. **Interpretation**:
   - Each \( \beta_i \) shows the effect of \( x_i \) on \( y \), assuming all other predictors are held constant.
   - \( R^2 \): Indicates the proportion of variance in \( y \) explained by the predictors (adjusted \( R^2 \) accounts for the number of predictors).

3. **Fitting the Model**:
   - Uses methods like Ordinary Least Squares (OLS) to minimize the sum of squared residuals (differences between actual and predicted \( y \)).

### Key Assumptions
Similar to Simple Linear Regression, with an additional assumption:
1. **Linearity**: The relationship between each \( x_i \) and \( y \) is linear.
2. **Independence**: Observations are independent.
3. **Homoscedasticity**: Constant variance of errors across all levels of predictors.
4. **Normality of Errors**: Errors are normally distributed with a mean of 0.
5. **No Multicollinearity**: Independent variables (\( x_1, x_2, \dots \)) are not highly correlated with each other, as multicollinearity can make coefficient estimates unstable or hard to interpret.

### Why Use Multiple Linear Regression?
- **Real-World Applicability**: Most outcomes depend on multiple factors, not just one (unlike Simple Linear Regression).
- **Control for Confounding**: It accounts for the effects of multiple variables, isolating the impact of each.
- **Flexibility**: Can handle continuous, categorical (via dummy variables), or transformed predictors.

### Example
To predict house prices, the model might be:

\[ \text{Price} = 30,000 + 150 \cdot \text{Size} + 20,000 \cdot \text{Bedrooms} + 10,000 \cdot \text{LocationScore} \]

Here, a 1-square-foot increase in size increases price by $150, a 1-bedroom increase adds $20,000, and a 1-unit increase in location score adds $10,000, assuming other variables are constant.

### Challenges
- **Multicollinearity**: High correlation between predictors (e.g., size and number of bedrooms) can distort coefficients. Detect using Variance Inflation Factor (VIF).
- **Overfitting**: Including too many predictors can fit noise rather than signal, reducing generalizability.
- **Assumption Violations**: Heteroscedasticity, non-linearity, or non-normal errors may require transformations or alternative models.



















## Q NO 5: What is polynomial regression, and how does it differ from linear regression ?

## ANS:

**Polynomial Regression** is a type of regression analysis that models the relationship between an independent variable (\( x \)) and a dependent variable (\( y \)) as an \( n \)-th degree polynomial. Unlike Simple Linear Regression, which assumes a straight-line relationship, Polynomial Regression allows for a curved, non-linear relationship by including higher-order terms (e.g., \( x^2 \), \( x^3 \)) in the model.

The equation for Polynomial Regression is:

\[ y = \beta_0 + \beta_1x + \beta_2x^2 + \beta_3x^3 + \dots + \beta_nx^n + \epsilon \]

- \( y \): Dependent variable (what you’re predicting)
- \( x \): Independent variable
- \( \beta_0, \beta_1, \dots, \beta_n \): Coefficients for each term
- \( x^2, x^3, \dots, x^n \): Higher-order terms of the independent variable
- \( \epsilon \): Error term
- \( n \): Degree of the polynomial (e.g., \( n=2 \) for quadratic, \( n=3 \) for cubic)

**Purpose**: To capture non-linear patterns in data, such as parabolic or more complex curves, that a straight line (Simple Linear Regression) cannot model effectively.

**Example**: Predicting crop yield (\( y \)) based on fertilizer amount (\( x \)) might show a quadratic relationship, where yield increases with fertilizer up to a point, then decreases due to over-fertilization. The model might be:

\[ \text{Yield} = \beta_0 + \beta_1 \cdot \text{Fertilizer} + \beta_2 \cdot \text{Fertilizer}^2 \]

### Key Differences from Linear Regression

1. **Linearity vs. Non-Linearity**:
   - **Simple Linear Regression**: Assumes a linear relationship between \( x \) and \( y \), modeled as a straight line (\( y = \beta_0 + \beta_1x \)).
   - **Polynomial Regression**: Models a non-linear relationship using polynomial terms, allowing the model to fit curves (e.g., parabolas for quadratic regression).

2. **Model Complexity**:
   - **Linear Regression**: Simpler, with only one independent variable term (\( x \)).
   - **Polynomial Regression**: More complex, as it includes higher-order terms (\( x^2, x^3, \dots \)), increasing flexibility but also computational complexity.

3. **Flexibility**:
   - **Linear Regression**: Limited to straight-line relationships, which may underfit data with non-linear patterns.
   - **Polynomial Regression**: Can fit a wider range of data patterns by adjusting the degree of the polynomial, making it more flexible for non-linear data.

4. **Risk of Overfitting**:
   - **Linear Regression**: Less prone to overfitting but may fail to capture complex relationships (underfitting).
   - **Polynomial Regression**: Higher-degree polynomials can overfit, fitting noise in the data rather than the true underlying pattern, especially with small datasets.

5. **Interpretability**:
   - **Linear Regression**: Coefficients (\( \beta_1 \)) are straightforward, representing the change in \( y \) per unit change in \( x \).
   - **Polynomial Regression**: Coefficients for higher-order terms (e.g., \( \beta_2 \)) are less intuitive, as they represent the effect of \( x^2 \), \( x^3 \), etc., making interpretation more complex.

### Key Assumptions (Similar to Linear Regression)
- The relationship between \( x \) and \( y \) can be approximated by a polynomial of degree \( n \).
- Observations are independent.
- Errors are normally distributed with a mean of 0.
- Variance of errors is constant (homoscedasticity), though this can be violated in practice.
- No extreme outliers that unduly influence the curve.

### Example to Illustrate Difference
Suppose you’re predicting house prices (\( y \)) based on square footage (\( x \)).
- **Linear Regression**: Might model \( \text{Price} = 50,000 + 200 \cdot \text{Size} \), assuming a constant price increase per square foot.
- **Polynomial Regression**: Might model \( \text{Price} = 50,000 + 300 \cdot \text{Size} - 0.05 \cdot \text{Size}^2 \), capturing a pattern where price increases with size but at a decreasing rate for very large houses.

### Practical Considerations
- **Choosing Polynomial Degree**: Higher degrees increase flexibility but risk overfitting. Use cross-validation or metrics like adjusted \( R^2 \) to select the appropriate degree.
- **Computational Complexity**: Polynomial Regression requires more computation, especially for higher degrees, due to additional terms.
- **Data Requirements**: Higher-degree polynomials need more data points to avoid overfitting and ensure stable estimates.

````python
# Q NO 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data:
# ● X = [1, 2, 3, 4, 5]
# ● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
# Plot the regression line over the data points.

# ANS:

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape for scikit-learn
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and fit the Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Get the slope and intercept
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope: {slope:.2f}, Intercept: {intercept:.2f}")
print(f"Equation: y = {slope:.2f}x + {intercept:.2f}")

# Predict values for the regression line
X_line = np.array([[0], [6]])  # Extend range slightly for plotting
Y_pred = model.predict(X_line)

# Calculate R-squared
r_squared = model.score(X, Y)
print(f"R-squared: {r_squared:.4f}")

# Plot the data points and regression line
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X_line, Y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Q NO 7: Fit a Multiple Linear Regression model on this sample data:
# ● Area = [1200, 1500, 1800, 2000]
# ● Rooms = [2, 3, 3, 4]
# ● Price = [250000, 300000, 320000, 370000]
# Check for multicollinearity using VIF and report the results.


#ANS:

# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

# Independent variables (X) and dependent variable (y)
X = df[['Area', 'Rooms']]
y = df['Price']

# Fit Multiple Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Get coefficients and intercept
intercept = model.intercept_
coef_area, coef_rooms = model.coef_
r_squared = model.score(X, y)

# Print regression results
print("Multiple Linear Regression Results:")
print(f"Intercept: {intercept:.2f}")
print(f"Coefficient for Area: {coef_area:.2f}")
print(f"Coefficient for Rooms: {coef_rooms:.2f}")
print(f"Equation: Price = {intercept:.2f} + {coef_area:.2f}*Area + {coef_rooms:.2f}*Rooms")
print(f"R-squared: {r_squared:.4f}")

# Calculate VIF for multicollinearity
X_with_const = pd.concat([X, pd.Series(1, index=X.index, name='const')], axis=1)  # Add constant for VIF
vif_data = pd.DataFrame()
vif_data['Variable'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X_with_const.values, i) for i in range(X.shape[1])]

# Print VIF results
print("\nVariance Inflation Factor (VIF) Results:")
print(vif_data)
print("\nInterpretation:")
for _, row in vif_data.iterrows():
    print(f"{row['Variable']}: VIF = {row['VIF']:.2f}")
    if row['VIF'] > 10:
        print(f"  - High multicollinearity (VIF > 10)")
    elif row['VIF'] > 5:
        print(f"  - Moderate multicollinearity (VIF > 5)")
    else:
        print(f"  - Low multicollinearity (VIF <= 5)")
````

````python
# Q NO 8: Implement polynomial regression on the following data:
# ● X = [1, 2, 3, 4, 5]
# ● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
# Fit a 2nd-degree polynomial and plot the resulting curve.

## ANS:

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape for scikit-learn
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create and fit Polynomial Regression model (degree 2)
degree = 2
polyreg = make_pipeline(PolynomialFeatures(degree), LinearRegression())
polyreg.fit(X, Y)

# Get coefficients and intercept
poly_features = PolynomialFeatures(degree)
X_poly = poly_features.fit_transform(X)
model = LinearRegression().fit(X_poly, Y)
intercept = model.intercept_
coef = model.coef_

# Print regression results
print("Polynomial Regression (Degree 2) Results:")
print(f"Intercept: {intercept:.2f}")
print(f"Coefficients: {coef[1]:.2f}*x + {coef[2]:.2f}*x^2")
print(f"Equation: y = {intercept:.2f} + {coef[1]:.2f}*x + {coef[2]:.2f}*x^2")
print(f"R-squared: {polyreg.score(X, Y):.4f}")

# Generate points for smooth curve
X_smooth = np.linspace(min(X), max(X), 100).reshape(-1, 1)  # 100 points for smooth curve
Y_smooth = polyreg.predict(X_smooth)

# Plot data points and polynomial curve
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X_smooth, Y_smooth, color='red', label='Polynomial Fit (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.grid(True)
plt.show()
````

## Q NO 10: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.

## ANS:

### Addressing Heteroscedasticity and Multicollinearity in House Price Prediction Model

As a data scientist at a real estate company, predicting house prices using features like area (e.g., square footage), number of rooms, and location (e.g., encoded as dummy variables for neighborhoods or proximity scores) is a common task. We typically start with a multiple linear regression model, such as:

\[ \text{Price} = \beta_0 + \beta_1 \cdot \text{Area} + \beta_2 \cdot \text{Rooms} + \beta_3 \cdot \text{Location} + \epsilon \]

However, detecting heteroscedasticity (non-constant variance of errors) and multicollinearity (high correlation among predictors) can bias inferences, inflate variances, and reduce model reliability. Below, I outline a step-by-step approach to diagnose (briefly, assuming initial detection), address these issues, and build a robust model. I'd use Python libraries like `statsmodels`, `scikit-learn`, and `matplotlib` for implementation.

#### Step 1: Confirm and Quantify the Issues
Before fixing, reconfirm the problems to guide targeted solutions:
- **For Multicollinearity**:
  - Calculate the Variance Inflation Factor (VIF) for each predictor using `statsmodels.stats.outliers_influence.variance_inflation_factor`. A VIF > 5 indicates moderate multicollinearity; >10 is severe. For example, area and rooms might be highly correlated (larger houses have more rooms), leading to high VIFs.
  - Plot a correlation matrix (using `pandas.corr()` and `seaborn.heatmap()`) to visualize pairwise correlations (e.g., if correlation between area and rooms > 0.7-0.8, it's a red flag).
- **For Heteroscedasticity**:
  - Generate a residuals plot (residuals vs. predicted prices) using `matplotlib`. Look for patterns like a funnel shape (widening spread at higher prices, common in real estate data where variance increases with price).
  - Perform a formal test like the Breusch-Pagan test (`statsmodels.stats.diagnostic.het_breuschpagan`) or White's test. A low p-value < 0.05 confirms heteroscedasticity.

If confirmed, proceed to remediation while monitoring other assumptions (e.g., linearity, normality of residuals via Q-Q plots).

#### Step 2: Address Multicollinearity
Multicollinearity doesn't bias coefficients but makes them unstable and hard to interpret (e.g., overestimating the impact of area while underestimating rooms). Here's how to fix it:
- **Remove or Combine Correlated Features**:
  - Drop one of the correlated variables if redundant (e.g., remove rooms if area captures most variance, based on domain knowledge—larger area often implies more rooms).
  - Create composite features, like "area per room" (`area / rooms`), to reduce redundancy.
- **Apply Dimensionality Reduction**:
  - Use Principal Component Analysis (PCA) from `sklearn.decomposition` to transform correlated features (area, rooms) into uncorrelated principal components. Fit PCA on the predictors, then use the components in regression. This is useful if location adds another layer of correlation (e.g., certain locations have larger homes).
- **Use Regularization Techniques**:
  - Switch to Ridge Regression (`sklearn.linear_model.Ridge`) or Lasso Regression (`sklearn.linear_model.Lasso`), which add a penalty (L2 or L1) to shrink coefficients and stabilize estimates. Tune the alpha parameter via cross-validation (`GridSearchCV`). For example:
    ```python
    from sklearn.linear_model import Ridge
    from sklearn.model_selection import GridSearchCV
    ridge = Ridge()
    params = {'alpha': [0.1, 1, 10, 100]}
    grid = GridSearchCV(ridge, params, cv=5)
    grid.fit(X, y)  # X: features, y: price
    ```
    Lasso can also perform feature selection by setting some coefficients to zero.
- **Collect More Data**: If feasible, gather additional samples to reduce the impact of multicollinearity, though this isn't always practical.

After these steps, recheck VIFs—aim for all < 5.

#### Step 3: Address Heteroscedasticity
Heteroscedasticity often arises in house price data because errors vary more for expensive homes (e.g., luxury properties have wider price fluctuations). It biases standard errors, leading to unreliable p-values and confidence intervals.
- **Transform the Dependent Variable**:
  - Apply a logarithmic transformation to price (`y = np.log(price)`) to stabilize variance, as prices are often right-skewed. Refit the model and check residuals. Other options: square root or Box-Cox transformation (`scipy.stats.boxcox`).
  - Interpretation changes (e.g., coefficients represent percentage changes in price), which is often more intuitive for real estate.
- **Use Weighted Least Squares (WLS)**:
  - Assign weights inversely proportional to variance (e.g., weight = 1 / predicted_price^2). Use `statsmodels.regression.linear_model.WLS`:
    ```python
    import statsmodels.api as sm
    weights = 1 / (model.predict(X)**2)  # Assuming variance proportional to predicted value
    wls_model = sm.WLS(y, sm.add_constant(X), weights=weights).fit()
    ```
- **Robust Standard Errors**:
  - Without changing the model, use heteroscedasticity-consistent (HC) standard errors via `statsmodels` (e.g., `cov_type='HC3'` in `OLS.fit()`). This corrects inference without fixing the underlying issue:
    ```python
    model = sm.OLS(y, sm.add_constant(X)).fit(cov_type='HC3')
    ```
- **Alternative Models**:
  - If transformations don't help, switch to Generalized Linear Models (GLM) or non-parametric methods like Random Forests (`sklearn.ensemble.RandomForestRegressor`), which don't assume constant variance.

Recheck with residuals plots and Breusch-Pagan test post-fixes—residuals should show random scatter with constant spread.

#### Step 4: Ensure Overall Model Robustness
After addressing the issues, validate and refine:
- **Cross-Validation and Performance Metrics**:
